```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy
import math
from test_functions import sinuses
%matplotlib notebook

def makeData (f):
    x = numpy.arange (0, 1, 0.01)
    y = numpy.arange (0, 1, 0.01)
    xgrid, ygrid = numpy.meshgrid(x, y)

    zgrid = f(xgrid, ygrid)
    return xgrid, ygrid, zgrid

fig = plt.figure(figsize=plt.figaspect(0.5))

axes = fig.add_subplot(2, 3, 1, projection='3d')
f = lambda x,y: sinuses([(1, 1), (1, 1, 2)]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(2, 3, 2, projection='3d')
f = lambda x,y: sinuses([(10, 100), (1, 1, 0)]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(2, 3, 3, projection='3d')
f = lambda x,y: sinuses([(100, 10), (1, 1, 0)]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(2, 3, 4, projection='3d')
f = lambda x,y: sinuses([(0, 0), (0, 0, 0), (0, 0, 0), (100, 100, 0)]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(2, 3, 5, projection='3d')
f = lambda x,y: sinuses([(0, 0), (1, 1, 0)]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(2, 3, 6, projection='3d')
f = lambda x,y: sinuses([(0, 0), (100, 1, 1)]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)
```
