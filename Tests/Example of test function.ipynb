```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import cm
import numpy
import math
%matplotlib notebook


def sinus(x, y):
    return (-1 * numpy.sin (x * math.pi / math.sqrt(2)) - 1 * numpy.sin (y * math.pi / math.sqrt(2)) + 27)**2+(-4 * numpy.sin (x * math.pi / math.sqrt(2)) - 5 * numpy.sin (y * math.pi / math.sqrt(2)) + 27)**1

def makeData (f):
    x = numpy.arange (0, 1, 0.01)
    y = numpy.arange (0, 1, 0.01)
    xgrid, ygrid = numpy.meshgrid(x, y)

    zgrid = f(xgrid, ygrid)
    return xgrid, ygrid, zgrid

x, y, z = makeData(sinus)

fig = pylab.figure()
axes = Axes3D(fig)

axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)
```
