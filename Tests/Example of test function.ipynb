# Graphics for tests functions

## Sinuses
Below images for sinuses functions (see [Description of tests functions properties](https://github.com/ASEDOS999/Optimization-Halving-The-Square/blob/master/Tests/Description.pdf) part 1.1).

```python
from test_functions import almost_polinomial as ap
import pylab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy
import math
%matplotlib notebook

def makeData_new (f, a, b, c, d):
    x = numpy.arange (a, b,  0.01)
    y = numpy.arange (c, d, 0.01)
    xgrid, ygrid = numpy.meshgrid(x, y)

    zgrid = f(xgrid, ygrid)
    return xgrid, ygrid, zgrid

fig = plt.figure(figsize=plt.figaspect(0.5))

axes = fig.add_subplot(3, 2, 1, projection='3d')
f = lambda x,y: ap([1, 1, 0, 0], [-1, 1, -1, 1]).calculate_function(x,y)
x, y, z = makeData_new(f, -1, 1, -1, 1)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 2, 2, projection='3d')
f = lambda x,y: ap([1, 1, 0, 0], [-5, 1, -5, 1]).calculate_function(x,y)
x, y, z = makeData_new(f, -5, 1, -5, 1)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 2, 3, projection='3d')
f = lambda x,y: ap([1, 1, 1, 1], [-1, 1, -1, 1]).calculate_function(x,y)
x, y, z = makeData_new(f, -1, 1, -1, 1)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 2, 4, projection='3d')
f = lambda x,y: ap([1, 1, 1, 5], [-2, 1, -2, 1]).calculate_function(x,y)
x, y, z = makeData_new(f, -2, 1, -2, 1)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 2, 5, projection='3d')
f = lambda x,y: ap([5, 1, 1, 1], [-2, 1, -2, 1]).calculate_function(x,y)
x, y, z = makeData_new(f, -2, 1, -2, 1)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 2, 6, projection='3d')
f = lambda x,y: ap([1, 1, 5, 5], [-2, 1, -2, 1]).calculate_function(x,y)
x, y, z = makeData_new(f, -2, 1, -2, 1)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)
```

## Almost Polimomial (Not Twice Differentiable Functions)
Below images for so called almost polimomial functions (see [Description of tests functions properties](https://github.com/ASEDOS999/Optimization-Halving-The-Square/blob/master/Tests/Description.pdf) part 1.2). This functions have Lipschitz continouos gradient and minimum at zero but have not the second derivative at zero.

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy
import math
from test_functions import sinuses
%matplotlib notebook

def makeData (f):
    x = numpy.arange (0, 1, 0.01)
    y = numpy.arange (0, 1, 0.01)
    xgrid, ygrid = numpy.meshgrid(x, y)

    zgrid = f(xgrid, ygrid)
    return xgrid, ygrid, zgrid

fig = plt.figure(figsize=plt.figaspect(0.5))

axes = fig.add_subplot(3, 3, 1, projection='3d')
f = lambda x,y: sinuses([(1, 1), (1, 1, 2)], [1.5, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 2, projection='3d')
f = lambda x,y: sinuses([(10, 100), (1, 1, 0)], [1.5, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 3, projection='3d')
f = lambda x,y: sinuses([(100, 10), (1, 1, 0)], [1.5, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 4, projection='3d')
f = lambda x,y: sinuses([(0, 0), (0, 0, 0), (0, 0, 0), (100, 100, 0)], [1.5, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 5, projection='3d')
f = lambda x,y: sinuses([(0, 0), (1, 1, 0)], [1.5, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 6, projection='3d')
f = lambda x,y: sinuses([(0, 0), (100, 1, 1)], [1.5, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 7, projection='3d')
f = lambda x,y: sinuses([(1, 1)], [1, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 8, projection='3d')
f = lambda x,y: sinuses([(1, 1)], [2, 2]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 9, projection='3d')
f = lambda x,y: sinuses([(1, 1)], [1, 1]).calculate_function(x,y)
x, y, z = makeData(f)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)
```
