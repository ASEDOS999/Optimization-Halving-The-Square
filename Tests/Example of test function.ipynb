# Graphics for tests functions

## Sinuses
Below images for sinuses functions (see [Description of tests functions properties](https://github.com/ASEDOS999/Optimization-Halving-The-Square/blob/master/Tests/Description.pdf) part 1.1).

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy
import math
from test_functions import sinuses
%matplotlib notebook

def makeData (f):
    x = numpy.arange (0, 1, 0.01)
    y = numpy.arange (0, 1, 0.01)
    xgrid, ygrid = numpy.meshgrid(x, y)

    zgrid = f(xgrid, ygrid)
    return xgrid, ygrid, zgrid

fig = plt.figure(figsize=plt.figaspect(0.5))

axes = fig.add_subplot(3, 3, 1, projection='3d')
f = lambda x,y: sinuses([(1, 1), (1, 1, 2)], [1.5, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
plt.xticks([])
plt.yticks([])
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 2, projection='3d')
f = lambda x,y: sinuses([(10, 100), (1, 1, 0)], [1.5, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
plt.xticks([])
plt.yticks([])
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 3, projection='3d')
f = lambda x,y: sinuses([(100, 10), (1, 1, 0)], [1.5, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
plt.xticks([])
plt.yticks([])
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 4, projection='3d')
f = lambda x,y: sinuses([(0, 0), (0, 0, 0), (0, 0, 0), (100, 100, 0)], [1.5, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
plt.xticks([])
plt.yticks([])
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 5, projection='3d')
f = lambda x,y: sinuses([(0, 0), (1, 1, 0)], [1.5, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
plt.xticks([])
plt.yticks([])
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 6, projection='3d')
f = lambda x,y: sinuses([(0, 0), (100, 1, 1)], [1.5, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
plt.xticks([])
plt.yticks([])
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 7, projection='3d')
f = lambda x,y: sinuses([(1, 1)], [1, 1.5]).calculate_function(x,y)
x, y, z = makeData(f)
plt.xticks([])
plt.yticks([])
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 8, projection='3d')
f = lambda x,y: sinuses([(1, 1)], [2, 2]).calculate_function(x,y)
x, y, z = makeData(f)
plt.xticks([])
plt.yticks([])
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

axes = fig.add_subplot(3, 3, 9, projection='3d')
f = lambda x,y: sinuses([(1, 1)], [1, 1]).calculate_function(x,y)
x, y, z = makeData(f)
plt.xticks([])
plt.yticks([])
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)
```

## Quadratic function

Below images for quadratic functions (see [Description of tests functions properties](https://github.com/ASEDOS999/Optimization-Halving-The-Square/blob/master/Tests/Description.pdf) part 1.2).

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np
import math
import random
from test_functions import quadratic_function as qf
%matplotlib notebook

def makeData (f, Q):
    x = np.arange (Q[0], Q[1], 0.01)
    y = np.arange (Q[2], Q[3], 0.01)
    xgrid, ygrid = np.meshgrid(x, y)

    zgrid = f(xgrid, ygrid)
    return xgrid, ygrid, zgrid

fig = plt.figure(figsize=plt.figaspect(0.5))

for i in range(9):
    axes = fig.add_subplot(3, 3, i + 1, projection='3d')
    param = np.random.uniform(-10, 10, 6)
    param[2] =  abs(param[2])
    f = qf(param)
    size_1, size_2 = random.uniform(0.5, 1), random.uniform(0.5, 1)
    x_1, y_1 = f.solution[0], f.solution[1]
    Q = [x_1 - (1-size_1), x_1 + (1+size_1), y_1 - (1-size_2), y_1 + (1+size_2)]
    func = lambda x,y: f.calculate_function(x,y)
    x, y, z = makeData(func, Q)
    plt.xticks([])
    plt.yticks([])

    axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)
```
